```python
import os
import pandas as pd
```

```python
# Path to the csv's
path_data_2015 = "Resources/raw_data/2015.csv"
path_data_2016 = "Resources/raw_data/2016.csv"
path_data_2017 = "Resources/raw_data/2017.csv"
path_data_2018 = "Resources/raw_data/2018.csv"
path_data_2019 = "Resources/raw_data/2019.csv"
path_countries_data="Resources/raw_data/countries.csv"
path_years_data="Resources/raw_data/years.csv"
path_ctry_rgn_data="Resources/raw_data/country_region.csv"
path_worldmap_data="Resources/raw_data/data_worldmap.csv"
```

```python
# Reading csv ->df
data_2015_df = pd.read_csv(path_data_2015)
data_2016_df = pd.read_csv(path_data_2016)
data_2017_df = pd.read_csv(path_data_2017)
data_2018_df = pd.read_csv(path_data_2018)
data_2019_df = pd.read_csv(path_data_2019)

countries_df=pd.read_csv(path_countries_data)
years_df=pd.read_csv(path_years_data)
ctry_rgn_df=pd.read_csv(path_ctry_rgn_data)
worldmap_data_df=pd.read_csv(path_worldmap_data)
```

```python
#Add region columns to 2017,2018 and 2019 df's
data_2017_df['region']=""
data_2018_df['region']=""
data_2019_df['region']=""
```

```python
# Drop the unnecessary columns from 2015,2016,2017
data_2015_df.drop(['Standard Error', 'Family'], axis=1, inplace=True)
data_2016_df.drop(['Lower Confidence Interval', 'Upper Confidence Interval','Family'], axis=1, inplace=True)
data_2017_df.drop(['Whisker.high', 'Whisker.low','Family',], axis=1, inplace=True)
```

```python
#Rename the columns
data_2015_df.columns = ['country', 'region', 'rank','score','gdp',
                        'health_life_expectancy','freedom','perceptions_of_corruption',
                       'generosity','dystopia_residual']

data_2016_df.columns = ['country', 'region', 'rank','score','gdp',
                        'health_life_expectancy','freedom','perceptions_of_corruption',
                       'generosity','dystopia_residual']

data_2017_df.columns = ['country','rank','score','gdp',
                        'health_life_expectancy','freedom','generosity',
                          'perceptions_of_corruption', 'dystopia_residual','region']

data_2018_df.columns = ['rank','country','score','gdp','social_support',
                        'health_life_expectancy','freedom','generosity',
                          'perceptions_of_corruption','region']

data_2019_df.columns = ['rank','country','score','gdp','social_support',
                        'health_life_expectancy','freedom','generosity',
                          'perceptions_of_corruption','region']
```

```python
# Rearrange columns
data_2015_df = data_2015_df[['rank','country','region','score','gdp',
                             'health_life_expectancy','freedom',
                             'perceptions_of_corruption','generosity','dystopia_residual']]

data_2016_df = data_2016_df[['rank','country','region','score','gdp',
                             'health_life_expectancy','freedom',
                             'perceptions_of_corruption','generosity','dystopia_residual']]

data_2017_df = data_2017_df[['rank','country','region','score','gdp',
                             'health_life_expectancy','freedom',
                             'perceptions_of_corruption','generosity','dystopia_residual']]

data_2018_df = data_2018_df[['rank','country','region','score','gdp',
                             'health_life_expectancy','freedom',
                             'perceptions_of_corruption','generosity','social_support']]

data_2019_df = data_2019_df[['rank','country','region','score','gdp',
                             'health_life_expectancy','freedom',
                             'perceptions_of_corruption','generosity','social_support']]
```

```python
# Populating Region column to 2017,2018 and 2019 df
ctry_rgn_dict = dict()
for index,row in ctry_rgn_df.iterrows() :
    country_name = row['country']
    region = row['region']
    ctry_rgn_dict[country_name] = region
ctry_rgn_dict
```

```python
# Function to populate region 2017,2018 and 2019 df
def add_region(df):
    for index,row in df.iterrows() :
        try:
            df.loc[index,'region'] = ctry_rgn_dict.get(row['country'])
        except:
            continue
```

```python
# Function call to populate the region 2017,2018 and 2019 df
add_region(data_2017_df)
add_region(data_2018_df)
add_region(data_2019_df)
```
